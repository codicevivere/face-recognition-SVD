#### Task 1:


Write a function rgbToGray to convert RGB image to Gray scale. A pixel grayvalue can be written as the linear combination of its R, G and B components.

$$G=0.2989 * R + 0.5870 * G + 0.1140 * B$$

You can write this expression in the form of dot product as well i.e.
\begin{equation*}
G=[0.2989\; 0.5870\; 0.1140]\left[
\begin{matrix}
R \\
G \\
B \\
\end{matrix}\right]\end{equation*}


```python
from skimage import img_as_int

def rgb2gray(im):
    """
        input: 
            im: a color image
        returns:
            oim: a gray scale image
    """
    return (im[:,:,0]*0.289)+(im[:,:,1]*0.5870)+(im[:,:,2]*0.1140)
```

```python
#Now call the function and convert the above image into gray-scale
nim=rgb2gray(im)
```

#### Task 2:
Write a function *colorHist* that accepts an  RGB color image and number of bins per colour as arguments and returns the histogram of color image. Provide a couple of examples with histograms. Note that you can implement this in two different ways, try thinking about the most logical ways. **Caution: Using large number of bins can make your system crash or hang.**
    

```python
def colorHist(im,nbins):
    """
        input: 
            im: a color image
            nbins: number of bins per colour in the histogram
        returns:
            hist: a list containing frequency of each bin
    """
    classwidth=(256/nbins)
    ans=np.zeros((nbins,nbins,nbins),dtype=np.int64)
    for i in arange(0,im.shape[0]):
        for j in arange(0,im.shape[1]):
            r=np.int(im[i,j,0]/classwidth)
            g=np.int(im[i,j,1]/classwidth)
            b=np.int(im[i,j,2]/classwidth)
            ans[r,g,b]+=1
    return ans 

hist=colorHist(im,8);
```
